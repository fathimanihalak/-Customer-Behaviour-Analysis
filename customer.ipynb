```python
from sqlalchemy import create_engine
import pandas as pd

# Step 1: Define your connection details
server = "DESKTOP-CTU0S4M\\SQLEXPRESS"
database = "Customer_behaviour"
driver = "ODBC Driver 17 for SQL Server"  # Ensure this driver is installed

# Step 2: Create connection string
connection_string = f"mssql+pyodbc://@{server}/{database}?driver={driver}&trusted_connection=yes"

# Step 3: Create SQLAlchemy engine
engine = create_engine(connection_string)

# Step 4: Test the connection
with engine.connect() as conn: 
    print("âœ… Successfully connected to SQL Server!")
```

```python
# Read your CSV file
df = pd.read_csv("customer_shopping_behavior.csv")

# Write to SQL Server table (creates a new table)
table_name = "CustomerData"
df.to_sql(table_name, engine, if_exists="replace", index=False)

print(f"âœ… Data successfully loaded into table '{table_name}' in database '{database}'.")
```
